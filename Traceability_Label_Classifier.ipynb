```python
from fastai.vision.all import *
from fastai.vision.widgets import *
```

```python
learn_inf = load_learner('export.pkl')
```

```python
btn_upload = widgets.FileUpload()
```

```python
btn_run = widgets.Button(description='Scan')
```

```python
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    if pred == "with label":
      lbl_pred.value = "Label Detected: PASS"
    else:
      lbl_pred.value = "Label Not Detected: PRODUCTION STOPPED"

btn_run.on_click(on_click_classify)
```

```python
file = open("usda-logo-color.png", "rb")
image = file.read()
```

```python
VBox([widgets.Label('USDA Label Looker 10000'), widgets.Image(value=image, format='png', width=150, height=200), widgets.Label('Select your image:'), 
      btn_upload, btn_run, out_pl, lbl_pred])
```
